# Exploring the Home Mortgage Market

#### Objective:
Our goal with this dataset is to explore the Home Mortgage market within the US to identify patterns the data on the basis of gender, race, income, property type, loan type, amount and location.

#### Data Brief: 
This dataset contains Mortgage applications for first lien, owner-occupied, 1-4 family homes under the Home Mortgage Disclosure Act (HMDA) between 2016 and 2017 (inclusive) reported by multiple financial institutions. For each record, we can learn about the loan, the property characteristics, the applicant demographics, and the lender. The dataset was modified to protect applicant and borrower privacy.

The original dataset comprises **13,023,011** records (null and non-null) and occupies **9.4 GB** in total.

#### Motivation:
Identifying the factors that affect mortgage attributes can help financial institutions and lenders make data-driven decisions and improve their lending practices. By understanding the key drivers of loan amounts, we can better assess credit risk, improve underwriting processes, and ultimately increase profitability.

##### Data sourced from Consumer Financial Protection Bureau (CFPB).  See __[here](https://www.consumerfinance.gov/data-research/hmda/historic-data/?geo=nationwide&records=first-lien-owner-occupied-1-4-family-records&field_descriptions=labels)__.

#### Report Summary:

The analysis of a dataset containing over 13 million mortgage applications between 2016 and 2017 under the Home Mortgage Disclosure Act (HMDA) by multiple financial institutions revealed several disparities and trends. Overall we observe that the dataset is biased towards males, with income disparities among different races. Loan amounts applied for tend to increase with income, but there is a widening gap in average loan amounts applied for by Asians and Whites. The Consumer Financial Protection Bureau provides the highest loan amounts to its applicants. Additionally, Debt-to-income ratios vary across states, with Californians having higher ratios than states on the east coast. These findings emphasize the need for addressing disparities, investigating lending practices, and economic factors to improve underwriting processes and increase profitability.

#### Distribution of Loans by Purpose

Q: What are the top loan purposes for the applications in the dataset?

A: The top loan purpose for the applications in the dataset is home purchase with a count of 7 million.

```python
# Create a temporary view of your DataFrame
data.createOrReplaceTempView(“data_view”)
# Run SQL query to group by ‘loan_purpose_name’, count, and order by count in descending order
loan_purpose_counts = spark.sql("""
                                SELECT loan_purpose_name, COUNT(*) as count
                                FROM data_view
                                GROUP BY loan_purpose_name
                                ORDER BY count DESC
                                """).collect()
# Use the result for plotting
purposes = [row[‘loan_purpose_name’] for row in loan_purpose_counts]
counts = [row[‘count’] for row in loan_purpose_counts]
plt.bar(purposes, counts)
plt.title("No. of Applications by Loan Purpose")
plt.xlabel('Loan Purpose')
plt.ylabel('Count')
plt.show()
```

#### Income of HEOPA Applicants

Q: What is the difference in average applicant income between HOEPA and non-HOEPA loan applications in the dataset?

A: The average income for non-HOEPA loans is higher, reaching over 100k, while for HOEPA loans, it stays below 80k. Therefore, the difference in average applicant income between HOEPA and non-HOEPA loan applications in the dataset is significant, with non-HOEPA loan applicants having a higher average income.

```python
import matplotlib.pyplot as plt
import pandas as pd
from pyspark.sql.functions import col

# Group the data by hoepa_status_name and compute the average applicant income
hoepa_applicant_income = (
    data.groupBy("hoepa_status_name")
        .agg({"applicant_income_000s": "mean"})
        .withColumnRenamed("avg(applicant_income_000s)", "Avg. Applicant Income (000's)")
)

# Convert to Pandas DataFrame and create a horizontal bar plot
hoepa_applicant_income_pd = hoepa_applicant_income.toPandas()
hoepa_applicant_income_pd.plot.barh(x="hoepa_status_name", y="Avg. Applicant Income (000's)", figsize=(6, 5))
plt.xlabel("Average Applicant Income (000's)")
plt.ylabel("HOEPA Status")
plt.title("Average Applicant Income by HOEPA Status")
plt.show()
```

#### Distribution of Income by Agencies

Q: What is the average income by agency name in the dataset,and how do they compare to each other?

A: The Federal Reserve System and Federal Deposit Insurance Corporation have an average applicant income of a little more than 100k. The Department of Housing and Urban Development has a little bit lower than 100k. The Office of Comptroller of the Currency has an average applicant income of more than 100k and almost reaches 120k. The National Credit Union Administration is the second-highest with more than 120k. The Consumer Financial Protection Bureau has the highest average applicant income at 140k.

```python
from pyspark.sql.functions import avg, when, sum, count

# Compute the average income and loan approval rate by agency name
agency_income_approval_rates = (
    data.groupBy("agency_name")
        .agg(avg("applicant_income_000s").alias("avg_income"),
             sum(when(col("action_taken_name") == "Loan originated", 1).otherwise(0)).alias("loan_approval_count"),
             count("*").alias("loan_count"))
        .withColumn("loan_approval_rate", col("loan_approval_count")/col("loan_count"))
        .select("agency_name", "avg_income", "loan_approval_rate")
)

import matplotlib.pyplot as plt

# Convert the Spark dataframe to a Pandas dataframe
agency_income_pd = agency_income_approval_rates.toPandas()

# Create a scatter plot of the average income by agency name
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(height=agency_income_pd["avg_income"], x=agency_income_pd["agency_name"])
ax.set_xlabel("Agency Name")
ax.set_ylabel("Average Income (000's)")
ax.set_title("Average Income by Agency Name")
plt.xticks(rotation=30)
plt.show()
```

#### Mean Income by Race

Q: What does the average income look like by race?

A: The graph presents the distribution of mean income by race, revealing significant disparities. Keeping aside the *Not applicable* and *Not Provided* categories, **Asian Americans** have the highest mean income of \\$135,000 per year, followed by **Whites** at \\$105,000 per year. **Native Americans**, **Black or African American**, and **Native Hawaiians** earn less than \\$100,000 per year on average. The data suggests that the income gap among different races persists and requires attention.

```python
# Cast Income to an Integer Type
income_by_race = data.withColumn("applicant_income_000s", F.col("applicant_income_000s").cast(IntegerType()))

# Group by Race to calculate the mean income for each race
income_by_race = income_by_race.groupBy("applicant_race_name_1").mean("applicant_income_000s").toPandas()

# Replace long value with a shorter version of it
income_by_race.loc[
    income_by_race["applicant_race_name_1"] == 
    "Information not provided by applicant in mail, Internet, or telephone application"
, "applicant_race_name_1"] = "Not Provided"

# Sort in Descending Order
income_by_race.sort_values(by = "avg(applicant_income_000s)", ascending = True, inplace = True)

# Rename Columns
income_by_race.columns = ["Race", "Average Income (in 000s)"]

# Plot the Bar Chart
#income_by_race.plot.bar(x = "Race", y = "Average Income (in 000s)", title = "Mean Income by Race");
fig = plt.figure(figsize = (10, 5))
 
# Creating the bar plot
plt.barh(income_by_race["Race"], income_by_race["Average Income (in 000s)"])

plt.ylabel("Race")
plt.xlabel("Average Income (in 000s)")
plt.title("Mean Income by Race")
plt.show();
```

#### Relationship Between Loan Amount and Applicant Income by Race

Q: How does the Loan Gap vary by Income and Race?

A: Based on the graph showing the variation of loan amount applied for versus the applicant's income, it appears that applicants tend to apply for higher amounts of loans as their income increases, which is an expected trend. Additionally, the graph suggests that there is a **widening** gap between the average loan amounts applied for by **Asians** and **Whites** as their income level increases. This could indicate disparities in lending practices or economic factors that need further investigation.

*Note:* To build this scatter plot, we will be sampling 0.5% of the data to have a visually coherent chart

```python
# Select only the relevant columns and remove unwanted rows
df = data\
        .sample(withReplacement = False, fraction = 0.005, seed = 3)\
        .select("loan_amount_000s", "applicant_income_000s", "applicant_race_name_1") \
        .filter(F.col("applicant_race_name_1").isin(["Asian", "Black or African American", "White"])) \
        .filter(~F.col("applicant_race_name_1").isin(\
             ["Joint", "Not applicable", "Not Hispanic or Latino", \
              "Information not provided by applicant in mail, Internet, or telephone application"]
        ))

# Convert the columns to numeric types and remove any rows with missing values
df = df.withColumn("loan_amount_000s", F.col("loan_amount_000s").cast("float")) \
       .withColumn("applicant_income_000s", F.col("applicant_income_000s").cast("float")) \
       .na.drop()

# Convert the PySpark DataFrame to a Pandas DataFrame for plotting
pdf = df.toPandas()

# Define a custom color scheme
colors = {
    "Asian": "#000D9E",
    "Black or African American": "#E69F00",
    "White": "#FF0000"
}

# Create a scatter plot of loan amount vs. applicant income, with color distinction based on applicant race
fig, ax = plt.subplots(figsize=(8, 6))
for race in pdf["applicant_race_name_1"].unique():
    race_df = pdf[pdf["applicant_race_name_1"] == race]
    ax.scatter(race_df["applicant_income_000s"], race_df["loan_amount_000s"], color=colors[race], label=race, s=4)

# Add a trend line to the plot
for race in pdf["applicant_race_name_1"].unique():
    race_df = pdf[pdf["applicant_race_name_1"] == race]
    slope, intercept, r_value, p_value, std_err = linregress(race_df["applicant_income_000s"], race_df["loan_amount_000s"])
    x = np.linspace(race_df["applicant_income_000s"].min(), race_df["applicant_income_000s"].max(), 100)
    y = slope * x + intercept
    ax.plot(x, y, color=colors[race], linestyle="--")

# Add axis labels and title
ax.set_xlabel("Applicant Income (in 000s)")
ax.set_ylabel("Loan Amount (in 000s)")
ax.set_title("Relationship Between Loan Amount and Applicant Income by Race")

# Create a legend for the color index and trend lines
handles, labels = ax.get_legend_handles_labels()
handles += [plt.plot([], [], color=colors[race], linestyle="--")[0] for race in pdf["applicant_race_name_1"].unique()]
labels += ["{} Trend Line".format(race) for race in pdf["applicant_race_name_1"].unique()]
ax.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Set x-axis limit
ax.set_xlim([0, 1000])
ax.set_ylim([0, 4500])

# Show the plot
plt.show()
```

#### Mean Loan Amount by Agencies

Q: Looking at Loan Agencies, how does the amounts they give out vary?

A: The graph shows the distribution of average loan amount given out by different agencies. ### Outlier Detection and Removal

The histogram of the income reveals a widespread skewness within it's values

```python
data\
    .withColumn('applicant_income_000s', F.col('applicant_income_000s').cast(DoubleType()))\
    .select("applicant_income_000s").toPandas().hist();
```

#### Remove Outliers

Here we use the Interquartile range of 1.5 to remove any outliers

```python
quantiles = data\
    .withColumn('applicant_income_000s', F.col('applicant_income_000s').cast(DoubleType()))\
    .approxQuantile('applicant_income_000s', [0.25, 0.75], 0.01)
q1 = quantiles[0]
q3 = quantiles[1]
iqr = q3 - q1

# Compute the lower and upper bounds of the range
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filter out any data point outside the range
data\
    .withColumn('applicant_income_000s', F.col('applicant_income_000s').cast(DoubleType()))\
    .filter(F.col('applicant_income_000s').between(lower_bound, upper_bound))\
    .select("applicant_income_000s").toPandas().hist();
```

#### Doing the same for Loan Amount

Here we perform the same activites for the Loan Amount column

```python
quantiles_loan_amount = data\
    .withColumn('loan_amount_000s', F.col('loan_amount_000s').cast(DoubleType()))\
    .approxQuantile('loan_amount_000s', [0.25, 0.75], 0.01)

q1_loan_amount = quantiles_loan_amount[0]
q3_loan_amount = quantiles_loan_amount[1]
iqr_loan_amount = q3_loan_amount - q1_loan_amount

# Compute the lower and upper bounds of the range
lower_bound_loan_amount = q1_loan_amount - 1.5 * iqr_loan_amount
upper_bound_loan_amount = q3_loan_amount + 1.5 * iqr_loan_amount
```

### Importing Packages

Import the packages for Machine Learning

```python
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder, StandardScaler
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.regression import LinearRegression
```

### Dividing columns into categorical and continuous columns
All the columns with string data type or less than 20 distinct values are detected as categorical columns, and numerical columns with more than 20 distinct values are continuous columns.

```python
# Detect categorical and numerical columns
categorical_columns = [field[0] for field in data_correct_type.dtypes if field[1] == "string" and field[0] != 'loan_amount_000s']
continuous_columns = [field[0] for field in data_correct_type.dtypes if field[1] != "string" and data_correct_type.select(field[0]).distinct().count() > 20 and field[0] != 'loan_amount_000s']
categorical_columns += [field[0] for field in data_correct_type.dtypes if field[1] != "string" and data_correct_type.select(field[0]).distinct().count() <= 20 and field[0] != 'loan_amount_000s']
categorical_columns = [column for column in categorical_columns if data_correct_type.select(column).distinct().count() > 1]
```

### Indexing and encoding
Indexing and one hot encoding all the categorical columns to make it compatible for linear regression model. Vector assembler combines all the feature columns into a single feature vector.

```python
# Index and one-hot encode categorical columns
indexers = [StringIndexer(inputCol=column, outputCol=f"{column}_index", handleInvalid="skip") for column in categorical_columns]
encoders = [OneHotEncoder(inputCol=f"{column}_index", outputCol=f"{column}_ohe") for column in categorical_columns]

# Assemble the features into a single vector
assembler = VectorAssembler(inputCols=[f"{column}_ohe" for column in categorical_columns] + continuous_columns, \
                            outputCol="features", handleInvalid="skip")
```

### Scaling and Pipeline
Adding standard scaler, creating a linear regression model and build the pipeline with the indexers, encoders, assembler, scaler and the model

```python
# Create the StandardScaler
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")

# Train the model
model = LinearRegression(featuresCol='scaledFeatures', labelCol='loan_amount_000s', maxIter=10, regParam=0.3, elasticNetParam=1)

# Build the pipeline
pipeline = Pipeline(stages=indexers + encoders + [assembler, scaler, model])
```

### Training and testing
Splitting the train and test data from a sample data, as we only really want to see the coefficients of the columns after the training is done, doing it on the sample should give us a good representation.

```python
# Split the data into training and testing sets
train_data, test_data = data_correct_type.sample(withReplacement = False, fraction = 0.3, seed = 3).randomSplit([0.8, 0.2], seed=4)

# Train the model
model = pipeline.fit(train_data)

# Make predictions
predictions = model.transform(test_data)

# Evaluate the model
evaluator = RegressionEvaluator(predictionCol='prediction', labelCol='loan_amount_000s', metricName='rmse')
rmse = evaluator.evaluate(predictions)

print("RMSE: ", rmse)
```

### Backtracking columns and coefficients
Using the metadata to find the coefficient indexes of the columns and matching them with coefficients. We are averaging the coefficients of the indexed and encoded columns (categorical columns).

```python
numeric_metadata = predictions.select("features").schema[0].metadata.get('ml_attr').get('attrs').get('numeric')
binary_metadata = predictions.select("features").schema[0].metadata.get('ml_attr').get('attrs').get('binary')

merge_list = numeric_metadata + binary_metadata 

colname_coff = [(row['name'], model.stages[-1].coefficients[row['idx']]) for row in merge_list]

all_columns = categorical_columns + continuous_columns
column_coefficient_map = {}
for column_name in all_columns:
    for name, coef in colname_coff:
        if column_name in name:
            if column_name not in column_coefficient_map:
                column_coefficient_map[column_name] = [abs(coef), 1]
            else:
                column_coefficient_map[column_name] = [column_coefficient_map[column_name][0] + abs(coef), column_coefficient_map[column_name][1]+1]
        
column_coefficient_map_scaled = {}
for column_name in column_coefficient_map:
    column_coefficient_map_scaled[column_name] = column_coefficient_map[column_name][0]/column_coefficient_map[column_name][1]
```

### Finding out most relevant columns
Narrowing down the relevant columns by creating a smaller model on the sample data and comparing coefficients of columns. 

```python

# Sort the dictionary by its values in descending order
sorted_dict = dict(sorted(column_coefficient_map_scaled.items(), key=lambda x: x[1], reverse=False))

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(8,10))

# Extract the keys and values from the sorted dictionary
keys = list(sorted_dict.keys())
values = list(sorted_dict.values())

# Create a horizontal bar chart
ax.barh(keys, values)

# Set the x-axis label
ax.set_xlabel('Coefficient Value')

# Show the plot
plt.show()
```

Our linear regression model shows that the primary determinants for loan amounts are:

- **Applicant Income**

    The applicant's income directly corresponds to their borrowing capacity, with higher incomes typically leading to larger loans.
    
- **State Name**

    The state can dictate the average loan size due to varying economic conditions and real estate prices.
    
- **Loan Purpose** and **Loan Type**

    The loan purpose and type also play a significant role, as certain purposes (like home purchases) and types (like FHA or VA loans) often involve larger sums.

- **The Year the loan was taken out**

    Temporal changes in economic conditions and lending policies, captured by the 'as of year' column, can influence loan amounts.
